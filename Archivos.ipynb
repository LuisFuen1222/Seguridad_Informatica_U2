```python
Los modos comunes son:

"r": Lectura (por defecto).
"w": Escritura (sobrescribe el archivo si ya existe).
"a": Añadir (agrega contenido al final del archivo si ya existe).
"b": Modo binario (se utiliza junto con otros modos, como "rb" o "wb").
```

```python
archivo = open("hola.txt","r")
resultado = archivo.read()

print(resultado)
```

```python
!pip install PyPDF2
import PyPDF2

pdf_file = open("CODE OF ETHIC FOR ENGINEER IDYGS71.pdf", "rb")

pdf_reader = PyPDF2.PdfReader(pdf_file)

for page in pdf_reader.pages:
    text = page.extract_text()
    print(text)

pdf_file.close()
```

```python
!pip install python-docx

import docx

doc = docx.Document('CODE OF ETHIC FOR ENGINEER.docx')

document_text = []

for paragraph in doc.paragraphs:
    document_text.append(paragraph.text)

full_text = '\n'.join(document_text)

print(full_text)
```
